```python
for test in set(df.columns.get_level_values(1)):
    print(test)
    not_days = []
    yes_days = []
    for day in set(df.columns.get_level_values(0)):
        if len(df[day][test].columns) > 0:
            yes_days.append(day)            
            columns = [x for x in df[day][test].columns.values]
        else:
            not_days.append(day)
    print(', '.join(not_days),' - no test')
    print(', '.join(yes_days),' - contain test')
    print(', '.join(columns))
    print('\n')
            
```
